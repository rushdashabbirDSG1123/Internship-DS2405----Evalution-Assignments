# Removing outliers

```python
#1. Drop the outliers and build the model check the performance

from scipy.stats import zscore
out_feature = df[['Open', 'High', 'Low', 'Volume']]
z= np.abs(zscore(out_feature))
z
#we taken list of values and not takine target variable.
#we can not findout outliers in the target variable or categorical variable 
# because in the categorical column we cant find out standardeviation mean, meadian 
#thats why in the categorical columns data are independent same as the target variable because target variable is also independent variable
# so we are not detecting the outliers from target and categorical variables


```

```python
#Threshold = 3

# Finding out the threshold value  99.7%  negitive -3 to +3
np.where(z>3)

#z = (datapoint/observed value -mean of the sample)/standerd deviation od the sample
```

# Find the Best random state

```python
from sklearn.model_selection  import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.metrics import r2_score
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.linear_model import LinearRegression
```

# Creating Train test Split

```python
# Finding the best random state
maxAccu = 0
maxRs =0
for i in range(1,200):
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .30, random_state=i)
    lr=LinearRegression()
    lr.fit(x_train,y_train)
    predlr = lr.predict(x_test)
    acc = r2_score(y_test,predlr)
    if acc>maxAccu:
        maxAccu=acc
        maxRs = i
print("Maximum r2 score is ",maxAccu,"On random state", maxRs)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = .25, random_state=maxRs)
```
